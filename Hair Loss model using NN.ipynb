```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv(r"C:\Users\Priyanshul\OneDrive\Desktop\Handwritten\hair_loss\Predict Hair Fall.csv")
```

```python
data=data.drop('Id',axis=1)
```

```python
categories=data.select_dtypes(include=['object','category']).columns.to_list()
```

```python
categories
```

```python
le=LabelEncoder()
for i in categories:
    data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
x=data.drop('Hair Loss',axis=1)
y=data["Hair Loss"]
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(11,)),
    Dense(32,activation='relu'),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(64,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```

```python
accuracy = model.evaluate(x_test, y_test)[1]
print(f'Model Accuracy: {accuracy * 100:.2f}%')
```
